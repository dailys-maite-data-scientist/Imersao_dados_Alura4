### Download dos dados

```python
url = 'https://raw.githubusercontent.com/dailys-maite-data-scientist/Imersao_dados_Alura4/main/dados.csv'
dados = pd.read_csv(url)
```

```python
dados.info()
```

```python
#Pegar o primeiro dado da coluna valor
dados["Valor"][0]
```

```python
#Dividir os dados ($ e valor)
dados["Valor"][0].split()
```

```python
#Criar um dataFrame com a separação de todas as sujeras
dados ["Valor"].str.split(expand = True)
```

```python
# O que tenho na coluna 0? É importante sempre explorar os dados para garantir que não vai dar problemas 
dados ["Valor"].str.split(expand = True)[0].unique()
```

```python
#Para saber quantos valores diferentes temos
len(dados ["Valor"].str.split(expand = True)[1].unique())
```

```python
dados ["Valor"].str.split(expand = True)[2].unique()
```

```python
#Selecionar mais de uma coluna
dados[["Valor", "Bairro"]]
```

```python
#Adicionar o dataframe criado ao nosso BD e colocar nome nas colunas
dados[["Moeda", "Valor_anuncio", "Tipo_anuncio"]] = dados ["Valor"].str.split(expand = True)
dados.head()
```

```python
# Filtrar os dados (Eliminar o tipo de anuncio que não é vacio
dados[dados["Tipo_anuncio"].isnull()]["Tipo_anuncio"].unique()
```

```python
#Ou seja ficamos só com os True (Null)
dados_vendas = dados[dados["Tipo_anuncio"].isnull()]
dados_vendas
```

```python
dados_vendas.info()
```

```python
#Eliminar o ponto da variável Valor_anuncio para ter a moeda no formato brasileiro
dados_vendas["Valor_anuncio"].str.replace(".","")
```

```python
# Convertir a coluna Valor_anuncio para float
dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)
```

```python
#Colocar o resultado no BD
dados_vendas["Valor_anuncio_float"] = dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)
```

```python
dados_vendas.info()
```

```python
dados_vendas["Valor_anuncio_float"].plot.hist()
```

```python
dados_vendas.describe()
```

```python
#Reduzir as casas decimais
pd.set_option("display.precision",2)
dados_vendas.describe()
```

```python
#Tirar notação cientifica
pd.set_option("display.float_format",lambda x: '%.2f' % x)
dados_vendas.describe()
```

```python
# Revisar todas as variaveis tanto as categoricas quanto as numericas
dados_vendas.describe(include = "all")
```

```python
# Exploração só de variaveis categóricas
dados_vendas[["Rua", "Bairro", "Cidade"]].describe()
```

```python
# Importar as bibliotecas mais usadas para plots
import seaborn as sns
#import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
#Plot com seaborn e matplotlib
sns.set()
plt.figure(figsize=(10, 8))
# O paramétro kde é para colocar a curva de aproximação no plot
ax = sns.histplot(data = dados_vendas, x = "Valor_anuncio_float", kde = True) 
ax.set_title("Histograma Valor Imóvel")
plt.xlim(-50, 10000000)
plt.show()
```
